```python
#import the required libraries
import pandas as pd
import numpy as np
import math

#Create the dataframe which has data that needs to be cleaned up
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
                    'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                    'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                    'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
                    '12. Air France', '"Swiss Air"']})

#Question 1: Fill in the missing numbers in the column FlightNumber in increasing order (increase by 10) and make the column an integer column (instead of a float)

#Iterage through the column FlightNumber using the length of dataframe as range
for i in range(0, len(df)):
    #If the value of current index is nan, update it using previous index value plus 10
    if math.isnan(df['FlightNumber'][i]):
        df['FlightNumber'][i] = df['FlightNumber'][i-1] + 10
                
#Convert the FlightNumber column from float to int
df['FlightNumber'] = df['FlightNumber'].astype('int64')  

#Display the updated dataframe
print("The dataframe after FlightNumber column clean up is:\n", df)
```

```python
#Question 2: Split each string on the underscore delimiter _ in column From_To to give a new temporary DataFrame with the correct values
#Assign the correct column names to this temporary DataFrame.

#Split each string in column From_To using map function and lambda
nlist = list((df['From_To'].astype('str').map(lambda x: x.split('_'))))

#Create temporary dataframe using the splitted list
tempdf = pd.DataFrame(nlist)

#Print the temporary dataframe before renaming the columns
print("Temporary dataframe before renaming the columns:\n", tempdf)

#Rename the columns to 'From' and 'To'
tempdf.columns = ['From', 'To']

#Print the temporary dataframe after renaming the columns
print("Temporary dataframe after renaming the columns:\n", tempdf)
```

```python
#Question 3: Standardise the strings so that only the first letter is uppercase (e.g."londON" should become "London".)

#Capitilize the first letter in 'From' column using map functiona and lambda
tempdf['From'] = tempdf['From'].astype('str').map(lambda x : x.capitalize())

#Capitilize the first letter in 'To' column using map functiona and lambda
tempdf['To'] = tempdf['To'].astype('str').map(lambda x : x.capitalize())

#Print the updated dataframe
print("Temporary dataframe after capitalizing the first letter is:\n",tempdf)
```

```python
#Question 4: Delete the From_To column from df and attach the temporary DataFrame
#Drop the column 'From_To' from DataFrame
df.drop(['From_To'], axis = 1, inplace = True)

#Print the dataframe after dropping column 'From_To'
print("The dataframe after dropping column 'From_To' is:\n", df)

#Attach the temporary dataframe to above dataframe
df = df.join(tempdf)

#Print the dataframe after attaching 'From' and 'To' columns from temporary dataframe
print("The dataframe after attaching 'From' and 'To' colummns is:\n", df)
```

```python
#Question 5: Expand the Series of lists into a DataFrame named delays, rename the columns delay_1, delay_2, etc. and replace the unwanted RecentDelays column in df with delays

#Expand the Series of lists 'RecentDelays' into a dataframe named delays
delays1 = pd.DataFrame(df['RecentDelays'].values.tolist())

#Print the dataframe 'delays'
print("The delays dataframe is:\n", delays1)

#Create a new list of column names after renaming the existing columns of the dataframe 'delays'
delays1.columns = ["delay_" + str(x+1) for x in delays1.columns.tolist()]

#Print the dataframe 'delays' after renaming the columns
print("The delays dataframe after renaming the columns is:\n", delays1)

#Drop the unwanted 'RecentDelays' column from the dataframe
df.drop(['RecentDelays'], axis = 1, inplace = True)

#Add the columns from 'delays' dataframe to the original dataframe
df = df.join(delays1)

#Print the dataframe with new columns
print("The dataframe after adding Delays columns is:\n", df)
```
